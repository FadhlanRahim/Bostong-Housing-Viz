```python
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(source,'lxml')
print(soup.title)
from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```

```python
#get the dataframe
file = source
df = pd.read_html(file)[0]

#delete unnasignned borough
df1 = df[df.Borough != 'Not assigned']

# Combining the neighbourhoods with same Postalcode and borough
df2 = df1.groupby(['Postal Code','Borough'], as_index=False).agg(lambda x: ','.join(x))
df2.rename(columns={'Postal Code':'PostalCode'},inplace=True)
df2
```

```python
latlondf = pd.read_csv('https://cocl.us/Geospatial_data')
```

```python
latlondf.rename(columns={'Postal Code':'PostalCode'},inplace=True)
latlondf
```

```python
df3 = pd.merge(df2, latlondf, on='PostalCode')
df3.head(12)
```

```python
#clustering toronto
df4 = df3[df3['Borough'].str.contains('Toronto')]
df4
```

```python
map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=12)

for lat,lng,borough,neighbourhood in zip(df4['Latitude'],df4['Longitude'],df4['Borough'],df4['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
    
map_toronto
```
